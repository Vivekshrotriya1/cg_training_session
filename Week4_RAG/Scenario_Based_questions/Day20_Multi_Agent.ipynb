#  Example 1: Sequential Multi-Agent System (Pipeline)
# Scenario

“A travel company has different employees (agents):

 Planner → decides steps

 Flight Agent → finds flights

 Weather Agent → checks weather

 Decision Agent → gives final answer”

```python
# ================================
# AGENT 1: PLANNER
# ================================
def planner_agent(user_query):
    print("\n[Planner Agent] Creating plan...")
    return ["flight", "weather", "decision"]


# ================================
# AGENT 2: FLIGHT AGENT
# ================================
def flight_agent():
    print("\n[Flight Agent] Fetching flights...")
    return [
        {"airline": "IndiGo", "price": 4500},
        {"airline": "Air India", "price": 5200}
    ]


# ================================
# AGENT 3: WEATHER AGENT
# ================================
def weather_agent():
    print("\n[Weather Agent] Checking weather...")
    return {"condition": "Clear", "temp": 28}


# ================================
# AGENT 4: DECISION AGENT
# ================================
def decision_agent(flights, weather):
    print("\n[Decision Agent] Making decision...")

    cheapest = min(flights, key=lambda x: x["price"])

    if weather["condition"] == "Rain":
        return "Avoid travel due to bad weather"

    return f"Book {cheapest['airline']} at ₹{cheapest['price']}"


# ================================
# MAIN MULTI-AGENT SYSTEM
# ================================
def travel_multi_agent(user_query):
    print("User Query:", user_query)

    plan = planner_agent(user_query)

    flights = None
    weather = None

    for step in plan:
        if step == "flight":
            flights = flight_agent()

        elif step == "weather":
            weather = weather_agent()

        elif step == "decision":
            result = decision_agent(flights, weather)

    return result


# RUN
response = travel_multi_agent("Plan my trip Delhi to Mumbai")
print("\nFinal Answer:", response)
```

#Scenario
“A hospital uses different employees (agents) to handle patient care in sequence.”


AGENT 1: Intake Agent (Planner)
- Collects patient symptoms and history
- Decides which steps are needed (tests, consultations, etc.)


AGENT 2: Diagnostic Agent
- Orders lab tests or scans
- Interprets results and identifies possible conditions

AGENT 3: Treatment Agent
- Suggests treatment options (medication, therapy, surgery)
- Considers patient preferences and medical guidelines

AGENT 4: Decision Agent
- Reviews all inputs (history, diagnostics, treatment options)
- Provides the final recommendation to the patient

```python
# ================================
# AGENT 1: INTAKE AGENT (PLANNER)
# ================================
def intake_agent(patient_query):
    print("\n[Intake Agent] Collecting patient data...")

    # Simulated patient input
    patient_data = {
        "symptoms": ["fever", "cough", "fatigue"],
        "history": "No major illness",
        "age": 30
    }

    print("Collected Data:", patient_data)

    # Plan what steps are needed
    plan = ["diagnosis", "treatment", "decision"]

    return patient_data, plan


# ================================
# AGENT 2: DIAGNOSTIC AGENT
# ================================
def diagnostic_agent(patient_data):
    print("\n[Diagnostic Agent] Analyzing symptoms...")

    symptoms = patient_data["symptoms"]

    # Simple rule-based diagnosis
    if "fever" in symptoms and "cough" in symptoms:
        diagnosis = "Possible Viral Infection"
        tests = ["Blood Test", "Chest X-ray"]
    else:
        diagnosis = "General Checkup Needed"
        tests = ["Basic Blood Work"]

    result = {
        "diagnosis": diagnosis,
        "recommended_tests": tests
    }

    print("Diagnostic Result:", result)
    return result


# ================================
# AGENT 3: TREATMENT AGENT
# ================================
def treatment_agent(diagnostic_result, patient_data):
    print("\n[Treatment Agent] Suggesting treatments...")

    diagnosis = diagnostic_result["diagnosis"]

    # Treatment suggestions
    if diagnosis == "Possible Viral Infection":
        treatment = [
            "Paracetamol",
            "Rest",
            "Hydration"
        ]
    else:
        treatment = ["General medication"]

    treatment_plan = {
        "diagnosis": diagnosis,
        "treatment_options": treatment,
        "notes": "Follow medical guidelines"
    }

    print("Treatment Plan:", treatment_plan)
    return treatment_plan


# ================================
# AGENT 4: DECISION AGENT
# ================================
def decision_agent(patient_data, diagnostic_result, treatment_plan):
    print("\n[Decision Agent] Finalizing recommendation...")

    final_decision = f"""
    Patient Summary:
    - Age: {patient_data['age']}
    - Symptoms: {', '.join(patient_data['symptoms'])}

    Diagnosis:
    - {diagnostic_result['diagnosis']}

    Recommended Tests:
    - {', '.join(diagnostic_result['recommended_tests'])}

    Treatment Plan:
    - {', '.join(treatment_plan['treatment_options'])}

    Advice:
    - Consult a doctor if symptoms persist
    """

    return final_decision


# ================================
# MAIN MULTI-AGENT SYSTEM
# ================================
def hospital_multi_agent(patient_query):
    print("Patient Query:", patient_query)

    patient_data, plan = intake_agent(patient_query)

    diagnostic_result = None
    treatment_plan = None

    for step in plan:
        if step == "diagnosis":
            diagnostic_result = diagnostic_agent(patient_data)

        elif step == "treatment":
            treatment_plan = treatment_agent(diagnostic_result, patient_data)

        elif step == "decision":
            result = decision_agent(
                patient_data,
                diagnostic_result,
                treatment_plan
            )

    return result


# ================================
# RUN SYSTEM
# ================================
response = hospital_multi_agent("I have fever and cough for 2 days")
print("\nFinal Recommendation:", response)
```

# Using groq api token

```python
from openai import OpenAI
import json
import os
from google.colab import userdata

# ================================
# GROQ / LLM SETUP
# ================================
api_key = userdata.get("groqapi")

client = OpenAI(
    api_key=api_key,   # ✅ Using secret key
    base_url="https://api.groq.com/openai/v1"
)

# ================================
# LLM CALL FUNCTION
# ================================
def call_llm(system_prompt, user_message):
    try:
        response = client.chat.completions.create(
            model="llama-3.3-70b-versatile",
            messages=[
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": user_message}
            ],
            temperature=0.3  # lower = more structured output
        )
        return response.choices[0].message.content.strip()

    except Exception as e:
        print("❌ LLM Error:", e)
        return ""


# ================================
# SAFE JSON PARSER (ROBUST)
# ================================
def safe_parse(text, default):
    try:
        # Extract JSON part if extra text exists
        start = text.find("[") if "[" in text else text.find("{")
        end = text.rfind("]") if "]" in text else text.rfind("}")

        if start != -1 and end != -1:
            text = text[start:end+1]

        return json.loads(text)

    except:
        print("⚠️ Parsing failed → using default")
        return default


# ================================
# AGENT 1: INTAKE AGENT
# ================================
def intake_agent(patient_query):
    print("\n[Intake Agent] Creating care plan...")

    reply = call_llm(
        system_prompt="""
Return ONLY a JSON list:
["diagnostics", "treatment", "decision"]
""",
        user_message=patient_query
    )

    plan = safe_parse(reply, ["diagnostics", "treatment", "decision"])
    print("Plan:", plan)
    return plan


# ================================
# AGENT 2: DIAGNOSTIC AGENT
# ================================
def diagnostic_agent(patient_query):
    print("\n[Diagnostic Agent] Running diagnosis...")

    reply = call_llm(
        system_prompt="""
Return ONLY JSON:
[
 {"condition": "name", "severity": "Mild/Moderate/Severe"}
]
""",
        user_message=patient_query
    )

    diagnostics = safe_parse(reply, [
        {"condition": "Flu", "severity": "Mild"}
    ])

    print("Diagnostics:", diagnostics)
    return diagnostics


# ================================
# AGENT 3: TREATMENT AGENT
# ================================
def treatment_agent(diagnostics):
    print("\n[Treatment Agent] Suggesting treatment...")

    reply = call_llm(
        system_prompt="""
Return ONLY JSON:
{
 "medication": "...",
 "rest": "...",
 "follow_up": "..."
}
""",
        user_message=str(diagnostics)
    )

    treatment = safe_parse(reply, {
        "medication": "Paracetamol",
        "rest": "Take proper rest",
        "follow_up": "Consult doctor if needed"
    })

    print("Treatment:", treatment)
    return treatment


# ================================
# AGENT 4: DECISION AGENT
# ================================
def decision_agent(patient_query, diagnostics, treatment):
    print("\n[Decision Agent] Final recommendation...")

    reply = call_llm(
        system_prompt="""
Give final recommendation in 2-3 simple sentences.
""",
        user_message=f"""
Patient: {patient_query}
Diagnosis: {diagnostics}
Treatment: {treatment}
"""
    )

    return reply


# ================================
# MAIN SYSTEM
# ================================
def hospital_multi_agent(patient_query):
    print("Patient Query:", patient_query)

    plan = intake_agent(patient_query)

    diagnostics = None
    treatment = None
    result = "No decision generated"

    for step in plan:
        if step == "diagnostics":
            diagnostics = diagnostic_agent(patient_query)

        elif step == "treatment":
            treatment = treatment_agent(diagnostics)

        elif step == "decision":
            result = decision_agent(patient_query, diagnostics, treatment)

    return result


# ================================
# RUN
# ================================
if __name__ == "__main__":
    response = hospital_multi_agent(
        "I have fever, sore throat and headache for 3 days"
    )

    print("\n===============================")
    print("✅ FINAL RECOMMENDATION:\n")
    print(response)
    print("===============================")
```

# Example 2: Manager–Worker Multi-Agent System
# Scenario

“Now instead of fixed flow, we introduce a Manager Agent
 that assigns tasks dynamically to worker

```python
def flight_agent():
    print("[Flight Agent] Working...")
    return [{"airline": "IndiGo", "price": 4500},
            {"airline": "Air India", "price": 5200}]


def weather_agent():
    print("[Weather Agent] Working...")
    return {"condition": "Clear", "temp": 28}


# ================================
# MANAGER AGENT
# ================================
def manager_agent(user_query):
    print("\n[Manager Agent] Analyzing task...")

    tasks = []

    if "trip" in user_query.lower():
        tasks = ["flight", "weather"]

    return tasks


# ================================
# EXECUTION
# ================================
def run_system(user_query):
    print("User Query:", user_query)

    tasks = manager_agent(user_query)

    results = {}

    for task in tasks:
        if task == "flight":
            results["flights"] = flight_agent()

        elif task == "weather":
            results["weather"] = weather_agent()

    # Final decision
    cheapest = min(results["flights"], key=lambda x: x["price"])

    return f"Manager Decision: Book {cheapest['airline']} at ₹{cheapest['price']}"


# RUN
response = run_system("Plan my trip")
print("\nFinal Answer:", response)
```

```python
from openai import OpenAI

# ================================
#   LLM SETUP (GROQ)
# ================================
import os
from openai import OpenAI
from google.colab import userdata

# ================================
#   LOAD API KEY FROM COLAB SECRETS
# ================================
# Make sure you added GROQ_API_KEY in Colab Secrets panel
api_key = userdata.get("groqapi")

# if not api_key:
#     raise ValueError("❌ GROQ_API_KEY not found in environment variables")

client = OpenAI(
    api_key=api_key,
    base_url="https://api.groq.com/openai/v1"
)

# ================================
# COMMON LLM FUNCTION
# ================================
def call_llm(role, message):
    try:
        response = client.chat.completions.create(
            model="llama-3.3-70b-versatile",
            messages=[
                {"role": "system", "content": role},
                {"role": "user", "content": message}
            ],
            temperature=0.4
        )
        return response.choices[0].message.content.strip()

    except Exception as e:
        print("❌ LLM Error:", e)
        return "No response due to API error"


# ================================
# ™ AGENT 1: CRISIS COORDINATOR
# ================================
def crisis_coordinator():
    print("\n[Crisis Coordinator] Broadcasting alert...")

    message = "Data breach detected in customer database. Immediate response required."

    print("Broadcast Message:", message)
    return message


# ================================
#   AGENT 2: IT SECURITY
# ================================
def it_security_agent(message):
    print("\n[IT Security Agent] Responding...")

    return call_llm(
        "You are an IT security expert. Give immediate technical actions.",
        message
    )


# ================================
#   AGENT 3: COMMUNICATIONS
# ================================
def communications_agent(message):
    print("\n[Communications Agent] Responding...")

    return call_llm(
        "You are a PR and communications expert. Suggest communication actions.",
        message
    )


# ================================
#   AGENT 4: FINANCE
# ================================
def finance_agent(message):
    print("\n[Finance Agent] Responding...")

    return call_llm(
        "You are a finance expert. Estimate financial actions and funding needs.",
        message
    )


# ================================
#   AGENT 5: LEGAL
# ================================
def legal_agent(message):
    print("\n[Legal Agent] Responding...")

    return call_llm(
        "You are a legal advisor. Suggest compliance and legal actions.",
        message
    )


# ================================
#   AGENT 6: HR
# ================================
def hr_agent(message):
    print("\n[HR Agent] Responding...")

    return call_llm(
        "You are an HR manager. Suggest employee-related actions.",
        message
    )


# ================================
#   AGENT 7: DECISION AGENT
# ================================
def decision_agent(responses):
    print("\n[Decision Agent] Creating final plan...")

    combined_input = "\n\n".join(
        [f"{agent.upper()}:\n{resp}" for agent, resp in responses.items()]
    )

    return call_llm(
        "You are a crisis management leader. Combine all inputs into a clear final response plan in bullet points.",
        combined_input
    )


# ================================
# MAIN SYSTEM
# ================================
def crisis_management_system():
    print("=== Crisis Management System (LLM Powered) ===")

    # Step 1: Broadcast
    message = crisis_coordinator()

    # Step 2: All agents respond
    responses = {
        "it": it_security_agent(message),
        "communications": communications_agent(message),
        "finance": finance_agent(message),
        "legal": legal_agent(message),
        "hr": hr_agent(message)
    }

    # Print responses
    print("\n--- Agent Responses ---")
    for k, v in responses.items():
        print(f"\n{k.upper()}:\n{v}")

    # Step 3: Final decision
    final = decision_agent(responses)

    return final


# ================================
# RUN
# ================================
if __name__ == "__main__":
    result = crisis_management_system()

    print("\n===============================")
    print("  FINAL CRISIS RESPONSE PLAN:\n")
    print(result)
    print("===============================")
```

#Scenario: Corporate Product Launch Broadcast
Imagine a company preparing to launch Product X in Q3. The Coordinator Agent (like a corporate program manager) sends out a broadcast announcement to all departments at once:
“Product X launch in Q3, target market North America, budget $5M.”


📢 Coordinator Agent (Broadcaster)
- Sends the launch announcement to all departments simultaneously.
- This is implemented in the code by the broadcast() function, which uses asyncio.gather() to run all agents in parallel.

📈 Marketing Agent
- Receives the broadcast.
- Responds with a marketing strategy: campaigns, channels, and positioning.
💰 Finance Agent
- Receives the broadcast.
- Responds with budget allocation and ROI forecasts.
🏭 Operations Agent
- Receives the broadcast.
- Responds with production and supply chain actions.
👩‍⚖️ Legal Agent
- Receives the broadcast.
- Responds with compliance checks and contract actions.
👩‍💼 HR Agent
- Receives the broadcast.
- Responds with staffing and training plans.

🧑‍⚖️ Decision Agent
- Collects all responses.
- Integrates them into a Final Corporate Launch Plan.
- In the code, this is the decision_agent() function that merges all outputs into one consolidated plan.

```python
# CORPORATE PRODUCT LAUNCH BROADCAST SYSTEM

import asyncio

# ================================
# WORKER AGENTS (Async)
# ================================

async def marketing_agent(broadcast):
    print("\n  [Marketing Agent] Received broadcast. Responding...")
    await asyncio.sleep(1)
    return {
        "agent": "Marketing",
        "actions": [
            "Launch digital campaigns on LinkedIn & Instagram",
            "Partner with North America influencers",
            "Position Product X as premium B2B solution"
        ],
        "status": "Marketing campaigns planned and channels confirmed."
    }

async def finance_agent(broadcast):
    print("\n  [Finance Agent] Received broadcast. Responding...")
    await asyncio.sleep(1)
    return {
        "agent": "Finance",
        "actions": [
            "Allocate $2M for marketing, $1.5M for ops, $1M for legal & HR",
            "Forecast ROI within 18 months",
            "Set up Q3 budget tracking dashboard"
        ],
        "status": "Budget allocated and ROI forecast prepared."
    }

async def operations_agent(broadcast):
    print("\n  [Operations Agent] Received broadcast. Responding...")
    await asyncio.sleep(1)
    return {
        "agent": "Operations",
        "actions": [
            "Scale production capacity by 40%",
            "Finalize North America distribution partners",
            "Set up inventory buffer for Q3 demand"
        ],
        "status": "Production scaled and supply chain finalized."
    }

async def legal_agent(broadcast):
    print("\n  [Legal Agent] Received broadcast. Responding...")
    await asyncio.sleep(1)
    return {
        "agent": "Legal",
        "actions": [
            "Register Product X trademark in North America",
            "Review distributor contracts and NDAs",
            "Ensure FTC compliance for marketing materials"
        ],
        "status": "Trademark filed and compliance checks completed."
    }

async def hr_agent(broadcast):
    print("\n  [HR Agent] Received broadcast. Responding...")
    await asyncio.sleep(1)
    return {
        "agent": "HR",
        "actions": [
            "Hire 30 sales reps for North America region",
            "Schedule product training for all teams",
            "Launch internal awareness campaign"
        ],
        "status": "Staffing plan ready and training scheduled."
    }


# ================================
# COORDINATOR AGENT (Broadcaster)
# ================================
def coordinator_agent():
    broadcast = "Product X launch in Q3, target market North America, budget $5M."
    print("\n[Coordinator Agent] Broadcasting launch announcement...")
    print(f" Broadcast : \"{broadcast}\"")
    print("  Sending to : Marketing, Finance, Operations, Legal, HR")
    return broadcast


# ================================
# BROADCAST (Async Parallel Execution)
# ================================
async def broadcast(broadcast_message):
    print("\n[System] All agents activated simultaneously via asyncio.gather()...")

    responses = await asyncio.gather(
        marketing_agent(broadcast_message),
        finance_agent(broadcast_message),
        operations_agent(broadcast_message),
        legal_agent(broadcast_message),
        hr_agent(broadcast_message)
    )

    return list(responses)


# ================================
# DECISION AGENT
# ================================
def decision_agent(responses):
    print("\n[Decision Agent] Integrating all department responses...")

    final_plan = " | ".join([r["status"] for r in responses])

    return {
        "total_agents_responded": len(responses),
        "final_launch_plan": final_plan,
        "verdict": "Product X Corporate Launch Plan activated across all departments."
    }


# ================================
# MAIN SYSTEM
# ================================
async def run_launch_system():

    print("   CORPORATE PRODUCT LAUNCH BROADCAST SYSTEM")

    # Step 1: Coordinator broadcasts
    message = coordinator_agent()

    # Step 2: All agents respond in parallel
    responses = await broadcast(message)

    # Step 3: Individual reports
    print("  DEPARTMENT RESPONSES")
    for r in responses:
        print(f"\n   {r['agent']} Agent")
        for action in r["actions"]:
            print(f"     • {action}")

    # Step 4: Decision agent consolidates
    decision = decision_agent(responses)

    print("  FINAL CORPORATE LAUNCH PLAN")
    print(f"\n  Departments Responded : {decision['total_agents_responded']}")
    print(f"\n   Plan : {decision['final_launch_plan']}")
    print(f"\n  {decision['verdict']}")

    return decision


# RUN
await run_launch_system()
```

# Real World Pipeline: Ai Research Team
Write an Comprenhensive market report on ev industryin india for 2025

```python
# ================================
# 🔐 IMPORTS
# ================================
import os
from openai import OpenAI

from google.colab import userdata
api_key= userdata.get('groqapi')

if not api_key:
    raise ValueError("❌ GROQ_API_KEY not found")

client = OpenAI(
    api_key=api_key,
    base_url="https://api.groq.com/openai/v1"
)

# ================================
# 🧠 SHARED MEMORY CLASS
# ================================
class SharedMemory:
    def __init__(self):
        self.data = {}

    def update(self, agent_name, result):
        self.data[agent_name] = result

    def get(self, key):
        return self.data.get(key, "")

    def get_all(self):
        return self.data


# ================================
# 🤖 LLM CALL FUNCTION (WITH TEMP)
# ================================
def call_llm(role, message, temperature):
    try:
        response = client.chat.completions.create(
            model="llama-3.3-70b-versatile",
            messages=[
                {"role": "system", "content": role},
                {"role": "user", "content": message}
            ],
            temperature=temperature
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        print("❌ LLM Error:", e)
        return "Error"


# ================================
# 🔎 SEARCH AGENT
# ================================
class SearchAgent:
    name = "search"

    def run(self, goal, memory):
        print("\n[Search Agent] Gathering EV data...")

        prompt = f"""
        Goal: {goal}

        Provide key data points about EV industry in India:
        - Market size
        - Growth trends
        - Major companies
        - Government policies
        """

        return call_llm(
            "You are a research assistant. Provide factual EV data.",
            prompt,
            temperature=0.2
        )


# ================================
# 📊 ANALYST AGENT
# ================================
class AnalystAgent:
    name = "analyst"

    def run(self, goal, memory):
        print("\n[Analyst Agent] Analyzing data...")

        search_data = memory.get("search")

        prompt = f"""
        Analyze the following EV data:

        {search_data}

        Provide insights:
        - Trends
        - Opportunities
        - Risks
        """

        return call_llm(
            "You are a business analyst. Interpret and summarize insights.",
            prompt,
            temperature=0.4
        )


# ================================
# ✍️ WRITER AGENT
# ================================
class WriterAgent:
    name = "writer"

    def run(self, goal, memory):
        print("\n[Writer Agent] Writing report...")

        analyst_data = memory.get("analyst")

        prompt = f"""
        Write a professional market report on:

        {goal}

        Based on:
        {analyst_data}

        Structure:
        - Introduction
        - Market Overview
        - Trends
        - Conclusion
        """

        return call_llm(
            "You are a professional report writer.",
            prompt,
            temperature=0.7
        )


# ================================
# ✅ QA AGENT
# ================================
class QAAgent:
    name = "qa"

    def run(self, goal, memory):
        print("\n[QA Agent] Reviewing report...")

        report = memory.get("writer")

        prompt = f"""
        Review the following report:

        {report}

        Check:
        - Accuracy
        - Clarity
        - Tone

        Improve if needed.
        """

        return call_llm(
            "You are a QA expert. Improve the report quality.",
            prompt,
            temperature=0.3
        )


# ================================
# 🧠 ORCHESTRATOR
# ================================
def orchestrator(goal):
    print("=== AI Research Team Pipeline ===")

    agents = [
        SearchAgent(),
        AnalystAgent(),
        WriterAgent(),
        QAAgent()
    ]

    memory = SharedMemory()

    for agent in agents:
        result = agent.run(goal, memory)
        memory.update(agent.name, result)

    return memory.get("qa")


# ================================
# 🚀 RUN SYSTEM
# ================================
if __name__ == "__main__":
    goal = "Write a comprehensive market report on EV industry trends in India for 2025."

    report = orchestrator(goal)

    print("\n===============================")
    print("📄 FINAL REPORT:\n")
    print(report)
    print("===============================")
```
