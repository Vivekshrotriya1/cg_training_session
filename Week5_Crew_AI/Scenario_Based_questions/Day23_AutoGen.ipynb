```python
# Scenario: Automated Business Report Assistant
# 🎯 Problem Statement
# In companies:
# - Managers spend hours compiling data from multiple sources (CRM, ERP, spreadsheets).
# - Reports often require manual formatting, chart creation, and consistency checks.
# - This slows down decision-making and introduces human error.
# 👉 Solution
# A multi-agent AutoGen system where:
# - 🤖 Assistant Agent drafts report templates and queries relevant data.
# - ⚙️ UserProxy Agent executes data pulls and formatting automatically.
# - 🔁 The system iterates until the report meets formatting, accuracy, and completeness standards.
# 🔄 Workflow
# Business Request → AI drafts report structure → Data fetched & formatted → Charts auto-generated → Validation loop → Final polished report


import autogen
import os


config_list = [
    {
        "model": "llama-3.3-70b-versatile",
        "api_key":api_key,
        "base_url": "https://api.groq.com/openai/v1"
    }
]

llm_config = {
    "config_list": config_list,
    "temperature": 0.2
}

report_agent = autogen.AssistantAgent(
    name="Report_Generator",
    llm_config=llm_config,
    system_message="""
You are a Business Report AI.

Your job:
1. Create structured reports
2. Ask for data analysis using Python
3. Ensure report includes:
   - Summary
   - KPIs
   - Charts
   - Insights
   - Recommendations

Always improve output iteratively.
"""
)

executor = autogen.UserProxyAgent(
    name="Executor",
    human_input_mode="ALWAYS",
    max_consecutive_auto_reply=6,
    code_execution_config={
        "work_dir": "reports",
        "use_docker": False
    }
)


critic = autogen.AssistantAgent(
    name="Critic",
    llm_config=llm_config,
    system_message="""
You review business reports.

Check:
- Missing insights
- Poor formatting
- Incorrect analysis

Suggest improvements clearly.
"""
)


groupchat = autogen.GroupChat(
    agents=[report_agent, executor, critic],
    messages=[],
    max_round=8
)

manager = autogen.GroupChatManager(
    groupchat=groupchat,
    llm_config=llm_config
)


task = """
Generate a business report for sales data:

Data: [12000, 15000, 10000, 18000, 13000]

Requirements:
- Calculate mean, median, mode
- Identify trends
- Generate a bar chart using matplotlib
- Provide business insights
- Give recommendations

Format:
1. Executive Summary
2. Data Analysis
3. Visualization
4. Insights
5. Recommendations
"""

executor.initiate_chat(manager, message=task)
```

#Scenario: AI Marketing Content & Validation System (Corporate Marketing Team)
🎯 Problem Statement
In companies:
- Marketing teams draft campaign content manually.
- Editors check for tone, compliance, and brand alignment.
- Testing across platforms (emails, social posts, ads) is manual and time-consuming.
👉 Solution
A multi-agent system where:
- ✍️ Content Creator Agent generates marketing copy, visuals, or ad text.
- 🔎 Validator Agent reviews for compliance, tone, and brand guidelines, then simulates deployment.
- 🔁 The loop continues until the content meets quality and compliance standards.
🔄 Workflow
Campaign Request → AI drafts content → Validator checks tone/compliance → Test deployment simulation → Iteration loop → Final approved campaign assets
🧹 Conflict Resolution
- Resolve conflicting brand guidelines (e.g., tone mismatch between product lines).
- Standardize formatting across channels (social, email, ads).
- Ensure package/library consistency for creative assets (e.g., Canva API vs. Adobe API).

```python
# ============================================
# 📦 IMPORTS
# ============================================
import os
import json
from openai import OpenAI


from openai import OpenAI

client = OpenAI(
    api_key=api_key,
    base_url="https://api.groq.com/openai/v1"
)

MODEL = "llama-3.3-70b-versatile"  # change if needed after testing


# ============================================
# 🧹 SAFE JSON PARSER
# ============================================
import re

def safe_json_parse(text):
    try:
        return json.loads(text)
    except:
        match = re.search(r'\{.*\}', text, re.DOTALL)
        if match:
            return json.loads(match.group())
        return {}


# ============================================
# ✍️ AGENT 1: CONTENT CREATOR
# ============================================
def content_creator(request):
    response = client.chat.completions.create(
        model=MODEL,
        temperature=0.7,
        messages=[
            {"role": "system", "content": "You are a marketing content creator."},
            {"role": "user", "content": f"""
Create marketing content for the following campaign:

{request}

Generate:
- email_campaign
- social_media_post
- ad_copy

Return STRICT JSON:
{{
  "email": "...",
  "social": "...",
  "ad": "..."
}}
"""}
        ]
    )

    return safe_json_parse(response.choices[0].message.content)


# ============================================
# 🔎 AGENT 2: VALIDATOR
# ============================================
def validator(content, guidelines):
    response = client.chat.completions.create(
        model=MODEL,
        temperature=0.3,
        messages=[
            {"role": "system", "content": "You are a strict marketing validator."},
            {"role": "user", "content": f"""
Validate this content based on guidelines:

CONTENT:
{json.dumps(content, indent=2)}

GUIDELINES:
{guidelines}

Check:
- Tone consistency
- Brand alignment
- Compliance issues
- Platform suitability

Return STRICT JSON:
{{
  "approved": true/false,
  "issues": ["..."],
  "suggestions": ["..."]
}}
"""}
        ]
    )

    return safe_json_parse(response.choices[0].message.content)


# ============================================
# 🔁 AGENT 3: ITERATIVE IMPROVER
# ============================================
def improve_content(content, feedback):
    response = client.chat.completions.create(
        model=MODEL,
        temperature=0.6,
        messages=[
            {"role": "system", "content": "You improve marketing content based on feedback."},
            {"role": "user", "content": f"""
Improve this content:

CONTENT:
{json.dumps(content, indent=2)}

FEEDBACK:
{json.dumps(feedback, indent=2)}

Return STRICT JSON in same format.
"""}
        ]
    )

    return safe_json_parse(response.choices[0].message.content)


# ============================================
# 🧹 CONFLICT RESOLUTION MODULE
# ============================================
def resolve_conflicts(guidelines):
    response = client.chat.completions.create(
        model=MODEL,
        temperature=0.2,
        messages=[
            {"role": "system", "content": "You resolve brand conflicts."},
            {"role": "user", "content": f"""
Resolve conflicts in these brand guidelines:

{guidelines}

Ensure:
- consistent tone
- unified messaging

Return clean unified guidelines.
"""}
        ]
    )

    return response.choices[0].message.content.strip()


# ============================================
# 🔄 MAIN PIPELINE
# ============================================
def run_pipeline():
    campaign_request = """
    Launch a new AI-powered productivity app targeting professionals.
    Focus on efficiency, automation, and time-saving.
    """

    brand_guidelines = """
    Tone should be professional but friendly.
    Avoid exaggerated claims.
    Keep messaging concise.
    Some teams prefer formal tone, others prefer casual tone.
    """

    print("\n🧹 Resolving guideline conflicts...\n")
    clean_guidelines = resolve_conflicts(brand_guidelines)
    print(clean_guidelines)

    print("\n✍️ Generating initial content...\n")
    content = content_creator(campaign_request)
    print(json.dumps(content, indent=2))

    # Iterative loop
    for i in range(3):
        print(f"\n🔎 Validation Round {i+1}...\n")
        feedback = validator(content, clean_guidelines)
        print(json.dumps(feedback, indent=2))

        if feedback.get("approved"):
            print("\n✅ Content Approved!\n")
            break

        print("\n🔁 Improving content...\n")
        content = improve_content(content, feedback)

    print("\n🚀 FINAL CONTENT:\n")
    print(json.dumps(content, indent=2))


# ============================================
# ▶️ RUN
# ============================================
if __name__ == "__main__":
    run_pipeline()
```

# Scenario: AI-Powered Dev Workflow (Corporate Engineering Team)
Problem Statement
 In companies:
 - Developers manually write and test code.
 - Reviewers check for bugs and compliance.
 - Testing and validation are slow and error-prone.

 👉 Solution
 A multi-agent system powered by Groq API where:
 - 👨‍💻 Coder Agent writes Python code for requested features.
 - 🔍 Reviewer Agent validates logic and ensures best practices.
 - 🧪 Tester Agent generates unit tests and executes them automatically.
 - 🔁 The loop continues until the code passes all tests and meets requirements.

 🔄 Workflow

 Feature Request → Product Manager breaks into tasks → Coder writes code → Tester validates with unit tests → Reviewer checks compliance → Iteration loop → Final approved codebase

 🧹 Conflict Resolution
 - Resolve conflicting dependencies (e.g., package version mismatches).
 - Standardize coding style across agents (PEP8 compliance).
 - Ensure Groq API integration avoids latency or model mismatch issues.

```python
import autogen
import os
from google.colab import userdata

# Groq API configuration
# Make sure you have stored your Groq API key in Colab Secrets as 'GROQ_API_KEY'
groq_api_key = userdata.get("VivekApi")

llm_config = {
    "config_list": [
        {
            "model": "llama-3.3-70b-versatile",   # Updated to a supported Groq model
            "api_key": groq_api_key,
            "base_url": "https://api.groq.com/openai/v1"  # Groq endpoint
        }
    ]
}

# Specialized agents
pm = autogen.AssistantAgent(
    name="ProductManager",
    llm_config=llm_config,
    system_message="You break down feature requests into tasks."
)

coder = autogen.AssistantAgent(
    name="Coder",
    llm_config=llm_config,
    system_message="You write Python code for given tasks."
)

tester = autogen.AssistantAgent(
    name="Tester",
    llm_config=llm_config,
    system_message="You write unit tests for code."
)

user = autogen.UserProxyAgent(
    name="User",
    human_input_mode="NEVER",
    code_execution_config={"work_dir": "project"}
)

# Group chat setup
groupchat = autogen.GroupChat(
    agents=[user, pm, coder, tester],
    messages=[],
    max_round=12,
    speaker_selection_method="auto"
)

# Manager orchestrates workflow
manager = autogen.GroupChatManager(
    groupchat=groupchat,
    llm_config=llm_config
)

# Start the group task
user.initiate_chat(manager,
    message="Build a simple REST API for a todo app")
```

# Scenario: AI-Powered HR Onboarding System (Corporate HR Team)
🎯 Problem Statement
In companies:
- HR teams manually prepare onboarding documents and checklists.
- Managers validate compliance with policies.
- Employees often face delays in receiving resources and training schedules.
👉 Solution
A multi-agent system where:
- 📝 HR Agent prepares onboarding documents, training schedules, and resource lists.
- ✅ Compliance Agent reviews documents for policy adherence and legal requirements.
- 📂 Resource Agent allocates accounts, devices, and access permissions.
- 🔁 The loop continues until all onboarding steps are validated and complete.
🔄 Workflow
Onboarding Request → HR Agent drafts documents → Compliance Agent validates → Resource Agent provisions accounts/devices → Iteration loop → Final approved onboarding package
🧹 Conflict Resolution
- Resolve conflicting policy requirements (e.g., different regional compliance rules).
- Standardize onboarding templates across departments.
- Ensure system avoids duplicate resource allocation (e.g., multiple accounts for same employee).

```python
# ============================================
# 📦 IMPORTS
# ============================================
import autogen
from google.colab import userdata

# ============================================
# 🔑 GROQ CONFIG
# ============================================
groq_api_key = userdata.get("VivekApi")

llm_config = {
    "config_list": [
        {
            "model": "llama-3.1-8b-instant",  # ⚡ FAST + LOW TOKEN,
            "api_key": groq_api_key,
            "base_url": "https://api.groq.com/openai/v1"
        }
    ],
    "temperature": 0.5
}

# ============================================
# 🤖 AGENTS
# ============================================

# 🧹 Conflict Resolver Agent
resolver = autogen.AssistantAgent(
    name="PolicyResolver",
    llm_config=llm_config,
    system_message="""
You resolve conflicts in HR policies.

Tasks:
- Remove contradictions (formal vs casual tone, region conflicts)
- Standardize rules across departments
- Output a clean unified policy
"""
)

# 📝 HR Agent
hr_agent = autogen.AssistantAgent(
    name="HRAgent",
    llm_config=llm_config,
    system_message="""
You are an HR onboarding specialist.

Responsibilities:
- Create onboarding documents
- Define training schedules
- List required resources

Return structured output:
{
  "documents": "...",
  "training": "...",
  "resources": [...]
}
"""
)

# ✅ Compliance Agent
compliance_agent = autogen.AssistantAgent(
    name="ComplianceAgent",
    llm_config=llm_config,
    system_message="""
You validate onboarding plans.

Check:
- Legal compliance (region-specific rules)
- Company policies
- Missing or incorrect steps

Return:
{
  "approved": true/false,
  "issues": [...],
  "suggestions": [...]
}
"""
)

# 📂 Resource Agent
resource_agent = autogen.AssistantAgent(
    name="ResourceAgent",
    llm_config=llm_config,
    system_message="""
You provision onboarding resources.

Tasks:
- Create email accounts
- Assign devices (laptop, etc.)
- Provide system access

Rules:
- Avoid duplicate allocation
- Ensure all required resources exist

Return:
{
  "provisioned": [...],
  "conflicts": [...]
}
"""
)

# 👤 User Proxy
user = autogen.UserProxyAgent(
    name="User",
    human_input_mode="NEVER",
    code_execution_config=False
)

# ============================================
# 💬 GROUP CHAT SETUP
# ============================================
groupchat = autogen.GroupChat(
    agents=[user, resolver, hr_agent, compliance_agent, resource_agent],
    messages=[],
    max_round=12,
    speaker_selection_method="auto"
)

manager = autogen.GroupChatManager(
    groupchat=groupchat,
    llm_config=llm_config
)

# ============================================
# 🚀 RUN PIPELINE
# ============================================
user.initiate_chat(
    manager,
    message="""
AI HR Onboarding Workflow:

Step 1: Resolve policy conflicts
Policies:
- Follow Indian labor laws
- Some departments prefer formal tone, others casual
- Avoid duplicate resource allocation
- Mandatory security training required

Step 2: HR Agent creates onboarding package

Step 3: Compliance Agent validates it

Step 4: Resource Agent provisions resources

Step 5: If issues exist → loop back and improve

Goal:
Generate a complete, compliant onboarding package for a Software Engineer in India.
"""
)
```
