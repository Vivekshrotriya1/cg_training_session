# Scenario: AI Content Automation for a Startup
 🏢 Problem Statement

 A startup wants to:

 Stay updated with latest AI trends

 Automatically create LinkedIn posts

 Save time on manual research & writing

 So we build a multi-agent AI system to automate this using CrewAI.

```python
!pip install crewai litellm langchain-community langchain-litellm

import os
os.environ["CREWAI_TRACING_ENABLED"] = "true"
from crewai import Agent, Task, Crew
from google.colab import userdata
from langchain_litellm import ChatLiteLLM # Updated import

# -----------------------------
# 1. Set Groq API Key
# -----------------------------
os.environ["GROQ_API_KEY"] = userdata.get("NewApi")

# ❌ REMOVE any OPENAI key completely
if "OPENAI_API_KEY" in os.environ:
    del os.environ["OPENAI_API_KEY"]

# -----------------------------
# 2. Define LLM (using ChatLiteLLM for Groq)
# -----------------------------
llm = ChatLiteLLM(model="groq/llama-3.3-70b-versatile")

# -----------------------------
# 3. Define Agents
# -----------------------------
researcher = Agent(
    role="Researcher",
    goal="Find latest AI news",
    backstory="Expert in AI trends",
    llm=llm, # Use the explicitly defined llm object
    verbose=True
)

writer = Agent(
    role="Writer",
    goal="Write LinkedIn post",
    backstory="Expert content creator",
    llm=llm, # Use the explicitly defined llm object
    verbose=True
)

# -----------------------------
# 4. Tasks
# -----------------------------
research_task = Task(
    description="Find 5 latest AI news trends",
    expected_output="List with summaries",
    agent=researcher
)

write_task = Task(
    description="Write LinkedIn post",
    expected_output="Final post",
    agent=writer,
    context=[research_task]
)

# -----------------------------
# 5. Crew
# -----------------------------
crew = Crew(
    agents=[researcher, writer],
    tasks=[research_task, write_task],
    verbose=True
)

# -----------------------------
# 6. Run
# -----------------------------
result = crew.kickoff()

print("\n✅ FINAL OUTPUT:\n")
print(result)
```

#Scenario: Automated Learning Hub for an EdTech Company
🎓 Problem Statement
An EdTech company wants to:
- Keep track of the latest educational technology innovations
- Automatically generate weekly blog posts for their website
- Reduce time spent on manual research and content drafting
👉 So we build a multi-agent AI system using CrewAI to automate this workflow:
- Trend Monitoring Agent → Continuously scans for new EdTech tools, teaching strategies, and AI-in-education updates
- Content Summarization Agent → Condenses research into digestible insights tailored for educators and learners
- Blog Writing Agent → Crafts engaging, SEO-friendly blog posts automatically
- Publishing Agent → Schedules and posts content directly to the company’s website

```python
# Allready installed libraries
# !pip install crewai litellm langchain-community langchain-litellm

# import os
# from crewai import Agent, Task, Crew
# from google.colab import userdata
# from langchain_litellm import ChatLiteLLM

# -----------------------------
# 1. API Setup (Groq)
# -----------------------------
os.environ["GROQ_API_KEY"] = userdata.get("NewApi")

if "OPENAI_API_KEY" in os.environ:
    del os.environ["OPENAI_API_KEY"]

# -----------------------------
# 2. LLM Setup
# -----------------------------
llm = ChatLiteLLM(
    model="groq/llama-3.3-70b-versatile",
    temperature=0.6
)

# -----------------------------
# 3. Agents
# -----------------------------

# 🔍 Trend Monitoring Agent
trend_monitor = Agent(
    role="EdTech Trend Analyst",
    goal="Identify latest innovations in educational technology, AI tools, and teaching strategies",
    backstory="Expert in EdTech research, constantly tracking global innovations in learning platforms and AI",
    llm=llm,
    verbose=True
)

# 🧠 Summarization Agent
summarizer = Agent(
    role="Content Summarizer",
    goal="Convert research into simple, clear insights for students and educators",
    backstory="Expert in simplifying complex ideas into easy-to-understand summaries",
    llm=llm,
    verbose=True
)

# ✍️ Blog Writing Agent
writer = Agent(
    role="Blog Writer",
    goal="Write engaging and SEO-optimized blog posts",
    backstory="Professional content creator specializing in educational blogs",
    llm=llm,
    verbose=True
)

# 📢 Publishing Agent
publisher = Agent(
    role="Publishing Manager",
    goal="Format, schedule, and prepare blog content for publishing",
    backstory="Handles blog publishing workflows and ensures content is ready for website upload",
    llm=llm,
    verbose=True
)

# -----------------------------
# 4. Tasks
# -----------------------------

# Task 1: Trend Monitoring
trend_task = Task(
    description="""
    Find latest EdTech trends including:
    - AI in education
    - New tools/platforms
    - Teaching innovations
    Provide at least 5 trends with brief explanations
    """,
    expected_output="List of 5+ EdTech trends with explanations",
    agent=trend_monitor
)

# Task 2: Summarization
summary_task = Task(
    description="""
    Convert the research into simple insights:
    - Easy to understand
    - Useful for students and educators
    - Highlight key takeaways
    """,
    expected_output="Simplified insights in bullet points",
    agent=summarizer,
    context=[trend_task]
)

# Task 3: Blog Writing
blog_task = Task(
    description="""
    Write a weekly blog post:
    - Catchy title
    - Introduction
    - Main content based on insights
    - SEO-friendly keywords
    - Conclusion + CTA
    """,
    expected_output="Full blog post (ready to publish)",
    agent=writer,
    context=[summary_task]
)

# Task 4: Publishing
publish_task = Task(
    description="""
    Prepare blog for publishing:
    - Format properly (headings, sections)
    - Suggest publishing schedule
    - Add tags and categories
    """,
    expected_output="Final formatted blog + publishing plan",
    agent=publisher,
    context=[blog_task]
)

# -----------------------------
# 5. Crew Setup
# -----------------------------
crew = Crew(
    agents=[trend_monitor, summarizer, writer, publisher],
    tasks=[trend_task, summary_task, blog_task, publish_task],
    verbose=True
)

# -----------------------------
# 6. Run
# -----------------------------
result = crew.kickoff()

print("\n✅ FINAL OUTPUT:\n")
print(result)
```

# Scenario: Smart Job Application Assistant
  Problem Statement

Students struggle to:

 Analyze job descriptions

 Match their skills

 Write tailored resumes & cover letters

 👉 So we build a multi-agent AI system that automates job application preparation.

 🤖 System Overview
 🔹 Agent 1: Job Analyzer

 Reads job description

 Extracts required skills

 🔹 Agent 2: Resume Optimizer

 Matches candidate skills with job

 Suggests improvements

 🔹 Agent 3: Cover Letter Writer

 Generates a personalized cover letter

 🔄 Workflow
 Job Description → Skill Analysis → Resume Optimization → Cover Letter Generation

```python



# !pip install crewai litellm

# import os
# from crewai import Agent, Task, Crew
# from google.colab import userdata

# -----------------------------
# 1. API Key Setup
# -----------------------------
os.environ["GROQ_API_KEY"] = userdata.get("NewApi")

# -----------------------------
# 2. Define Agents
# -----------------------------
job_analyzer = Agent(
    role="Job Analyzer",
    goal="Extract key skills and requirements from job descriptions",
    backstory=(
        "You are an HR expert who understands job postings and identifies "
        "important skills, qualifications, and expectations."
    ),
    llm="groq/llama-3.3-70b-versatile",
    verbose=True
)

resume_optimizer = Agent(
    role="Resume Optimizer",
    goal="Align candidate resume with job requirements",
    backstory=(
        "You are a career coach who improves resumes by matching them "
        "with job descriptions and highlighting relevant skills."
    ),
    llm="groq/llama-3.3-70b-versatile",
    verbose=True
)

cover_letter_writer = Agent(
    role="Cover Letter Writer",
    goal="Write a personalized cover letter",
    backstory=(
        "You are a professional writer who crafts compelling cover letters "
        "tailored to specific job roles."
    ),
    llm="groq/llama-3.3-70b-versatile",
    verbose=True
)

# -----------------------------
# 3. Input Data (Scenario)
# -----------------------------
job_description = """
We are looking for a Data Analyst with experience in Python, SQL, Power BI,
and data visualization. The candidate should be able to analyze datasets,
create dashboards, and communicate insights effectively.
"""

candidate_resume = """
I am a BTech student skilled in Python and basic SQL.
I have created Power BI dashboards and completed projects on data analysis.
"""

# -----------------------------
# 4. Tasks
# -----------------------------
analysis_task = Task(
    description=f"""
    Analyze the following job description and extract:
    - Required skills
    - Key responsibilities
    - Tools/technologies

    Job Description:
    {job_description}
    """,
    expected_output="Structured list of skills, responsibilities, and tools",
    agent=job_analyzer
)

resume_task = Task(
    description=f"""
    Compare the candidate resume with the job requirements and:
    - Identify skill gaps
    - Suggest improvements
    - Optimize the resume

    Resume:
    {candidate_resume}
    """,
    expected_output="Improved resume with suggestions",
    agent=resume_optimizer,
    context=[analysis_task]
)

cover_letter_task = Task(
    description="""
    Create a personalized cover letter using:
    - Job requirements
    - Candidate resume
    - Improvements suggested

    Make it professional and engaging.
    """,
    expected_output="Final cover letter",
    agent=cover_letter_writer,
    context=[analysis_task, resume_task]
)

# -----------------------------
# 5. Crew Setup
# -----------------------------
crew = Crew(
    agents=[job_analyzer, resume_optimizer, cover_letter_writer],
    tasks=[analysis_task, resume_task, cover_letter_task],
    verbose=True,
    tracing=True   # good for demo
)

# -----------------------------
# 6. Run
# -----------------------------
result = crew.kickoff()

print("\n🚀 FINAL OUTPUT:\n")
print(result)
```

# Problem Statement
Enterprise sales teams often struggle to:
- Analyze client requirements from RFPs (Requests for Proposals)
- Match company offerings to client needs
- Draft tailored proposals that stand out
- Ensure consistency across documents
👉 So we build a multi-agent AI system that automates sales proposal preparation.

🤖 System Overview
- 🔹 Agent 1: RFP Analyzer
- Reads client’s request for proposal
- Extracts key requirements, budget constraints, and evaluation criteria
- 🔹 Agent 2: Solution Mapper
- Matches company products/services to client needs
- Suggests tailored value propositions and differentiators
- 🔹 Agent 3: Proposal Writer
- Generates a polished, client-specific proposal
- Ensures alignment with brand tone and compliance standards

🔄 Workflow
RFP Document → Requirement Analysis → Solution Mapping → Proposal Generation

```python

import json
from openai import OpenAI


from google.colab import userdata
api_key= userdata.get('VivekApi')
client = OpenAI(
    api_key=api_key,
    base_url="https://api.x.ai/v1"
)


def analyze_rfp(rfp_text):
    try:
        response = client.chat.completions.create(
            model="grok-2-latest",
            temperature=0.3,
            messages=[
                {"role": "system", "content": "You are an expert business analyst extracting structured insights from RFP documents."},
                {"role": "user", "content": f"""
Extract the following details from the RFP:

- client_name
- requirements (list)
- budget
- timeline
- evaluation_criteria (list)

Return STRICT JSON only.

RFP:
{rfp_text}
"""}
            ]
        )

        content = response.choices[0].message.content.strip()
        return json.loads(content)

    except Exception as e:
        print("❌ Error in Analyzer:", e)
        return {}

# ==============================
# 🤖 AGENT 2: SOLUTION MAPPER
# ==============================
def map_solution(analysis_json):
    try:
        response = client.chat.completions.create(
            model="grok-2-latest",
            temperature=0.4,
            messages=[
                {"role": "system", "content": "You map business requirements to company solutions."},
                {"role": "user", "content": f"""
Based on this client analysis:

{json.dumps(analysis_json, indent=2)}

Match:
- each requirement → best solution
- include value proposition

Return STRICT JSON in format:
{{
  "solutions": [
    {{
      "requirement": "...",
      "solution": "...",
      "value": "..."
    }}
  ]
}}
"""}
            ]
        )

        content = response.choices[0].message.content.strip()
        return json.loads(content)

    except Exception as e:
        print("❌ Error in Mapper:", e)
        return {}

# ==============================
# 🤖 AGENT 3: PROPOSAL WRITER
# ==============================
def generate_proposal(mapping_json, analysis_json):
    try:
        response = client.chat.completions.create(
            model="grok-2-latest",
            temperature=0.7,
            messages=[
                {"role": "system", "content": "You are a professional proposal writer."},
                {"role": "user", "content": f"""
Write a professional sales proposal using:

Client Info:
{json.dumps(analysis_json, indent=2)}

Solutions:
{json.dumps(mapping_json, indent=2)}

Include:
- Title
- Introduction
- Problem Understanding
- Proposed Solution
- Value Proposition
- Timeline
- Budget (if available)
- Closing Statement

Make it formal and client-specific.
"""}
            ]
        )

        return response.choices[0].message.content.strip()

    except Exception as e:
        print("❌ Error in Writer:", e)
        return "Proposal generation failed."

# ==============================
# 📄 OPTIONAL: SAVE TO FILE
# ==============================
def save_proposal(proposal_text):
    try:
        with open("final_proposal.txt", "w", encoding="utf-8") as f:
            f.write(proposal_text)
        print("✅ Proposal saved as final_proposal.txt")
    except Exception as e:
        print("❌ Error saving file:", e)

# ==============================
# 🔄 MAIN PIPELINE
# ==============================
def run_pipeline():
    # Sample RFP (you can replace with PDF text later)
    rfp_text = """
    ABC Corp is looking for a cloud migration solution along with strong data security.
    Budget: $50,000 to $100,000
    Timeline: 3 months
    Evaluation criteria include cost efficiency, prior experience, and 24/7 support.
    """

    print("\n🔍 STEP 1: Analyzing RFP...\n")
    analysis = analyze_rfp(rfp_text)
    print(json.dumps(analysis, indent=2))

    print("\n🧠 STEP 2: Mapping Solutions...\n")
    mapping = map_solution(analysis)
    print(json.dumps(mapping, indent=2))

    print("\n✍️ STEP 3: Generating Proposal...\n")
    proposal = generate_proposal(mapping, analysis)
    print(proposal)

    # Save output
    save_proposal(proposal)


# ==============================
# ▶️ RUN
# ==============================
if __name__ == "__main__":
    run_pipeline()
```
