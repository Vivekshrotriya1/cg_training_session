#  Scenario: University Library Search
Imagine you’re building a smart search system for a university library.
 - Problem: Students often type queries like “fast dog running” or “canine sprinting”, but
 the library catalog has books titled “The dog runs fast”. Traditional keyword search misses these connections.
 - Solution: Use embeddings to capture semantic meaning instead of exact words.

```python



from sentence_transformers import SentenceTransformer



model_name = "all-MiniLM-L6-v2"
model = SentenceTransformer(model_name)

texts = [
    "The dog runs fast",
    "A quick canine sprints",
    "Python is a programming language"
]

embeddings = model.encode(texts)

print(f"Model: {model_name}")
print(f"Embedding shape: {embeddings.shape}")
print(f"First vector sample: {embeddings[0][:10]}")
```

# Cosine Similarity

```python
# Simple cosine similarity example
import numpy as np

def cosine_similarity(a, b):
    return np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b))

a = np.array([0.2,0.8,-0.3,0.5])
b = np.array([0.1,0.9,-0.2,0.4])

print(cosine_similarity(a,b))
```

#Scenario: Healthcare Chatbot
– Patient Symptom Matching Imagine you’re designing a healthcare chatbot that helps patients find possible conditions based on their symptoms. Both patients and conditions are represented as vectors of features (e.g., fever, cough, fatigue, headache).

Patient’s symptom vector: [0.2,0.8,-0.3,0.5] → mild fever, strong cough, no fatigue, moderate headache.
Condition’s symptom profile vector: [0.1,0.9,-0.2,0.4] → mild fever, strong cough, little fatigue, moderate headache. The chatbot uses cosine similarity to measure how closely the patient’s symptoms align with the condition’s profile

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

patient=np.array([[0.2, 0.8, -0.3, 0.5]])
condition=np.array([[0.1, 0.9, -0.2, 0.4]])
similarity=cosine_similarity(patient, condition)

print("Patient Symptom Vector:", patient)
print("Condition Symptom Vector:", condition)
print("Cosine Similarity:", similarity[0][0])
```

#Simple Euclidean Distance (L2 Distance)

```python
import numpy as np

a = np.array([1,2])
b = np.array([2,6])

distance = np.linalg.norm(a-b)

print(distance)
```

#Dot Product (Fastest Similarity Method)

```python

# Compute dot product similarity
a_norm = a / np.linalg.norm(a)
b_norm = b / np.linalg.norm(b)

np.dot(a_norm,b_norm)
```

# Scenario: E‑Commerce Product Recommendations

Imagine you’re building a smart recommendation engine for an online store.

Problem: Customers often search with phrases like “comfortable running shoes”, but the catalog has items titled “lightweight sneakers for jogging”. Traditional keyword search struggles to connect these.
Solution: Use embeddings to capture semantic meaning of product descriptions and customer queries.

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

#  Customer query embedding (semantic meaning of "comfortable running shoes")
query_embedding = np.array([0.1, 0.8, 0.5])

#  Product catalog embeddings (semantic meaning of product descriptions)
product_embeddings = {
    "lightweight sneakers for jogging": np.array([0.2, 0.7, 0.6]),
    "formal leather shoes": np.array([0.9, 0.1, 0.2]),
    "sandals for summer": np.array([0.3, 0.4, 0.5])
}

#  Compute similarity between query and each product
for product, embedding in product_embeddings.items():
    similarity = np.dot(query_embedding, embedding)  # fast similarity check
    cosine_sim = cosine_similarity([query_embedding], [embedding])[0][0]  # normalized similarity
    print(f"{product} → Dot Product: {similarity:.3f}, Cosine Similarity: {cosine_sim:.3f}")

best_match = max(product_embeddings.items(), key=lambda x: np.dot(query_embedding, x[1]))
print("Recommended Product:", best_match[0])
```

# Scenario: Online Learning Platform Recommendations

Imagine you’re building a smart recommendation engine for an online education platform.

Problem: Students often search with phrases like “beginner-friendly Python tutorials”, but the course catalog has items titled “Introduction to Programming with Python”. Traditional keyword search struggles to connect these because the words don’t match exactly.
Solution: Use embeddings to capture the semantic meaning of both student queries and course descriptions. By comparing embeddings with dot product or cosine similarity, the system can recommend the most relevant courses even when the wording is different

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

model=SentenceTransformer('all-MiniLM-L6-v2')

courses=[
    "Introduction to Programming with Python",
    "Advanced Data Structures in Java",
    "Web Development with HTML CSS and JavaScript",
    "Machine Learning Fundamentals"
]

query = "beginner friendly Python tutorials"
course_embeddings = model.encode(courses)
query_embedding = model.encode([query])
similarity = cosine_similarity(query_embedding, course_embeddings)

print("Similarity scores:", similarity)
```

#Scenario: Student Course Recommendations

Imagine you’re running an online learning platform. Students type queries like “beginner-friendly Python tutorials”, but your catalog has courses titled “Introduction to Programming with Python”. Traditional keyword search might miss the match, but embeddings + vector search (Chroma locally or Pinecone in the cloud) can connect them semantically.

🧩 Teaching Exercise Flow

Documents (Courses)
"Python is a high-level programming language"
"Machine learning models need training data"
"Dogs are loyal and friendly animals"
"Cats are independent and curious pets"
Embedding Model
Converts each course description into a vector (numbers capturing meaning).
Indexing
Chroma (local): Great for classroom demos, no API key needed.
Pinecone (cloud): Scales to billions of items, perfect for production.
Query
Student asks: “What animals make good companions?”
Embedding of query is compared against course embeddings.
Results
Top matches:
Cats are independent and curious pets
Dogs are loyal and friendly animals

```python
!pip install chromadb
import chromadb
from sentence_transformers import SentenceTransformer

# Load embedding model
model = SentenceTransformer("all-MiniLM-L6-v2")

# Create Chroma client and collection
client = chromadb.Client()  # or chromadb.PersistentClient(path="./db")
collection = client.create_collection("docs2")

# Sample documents
documents = [
    "Python is a high-level programming language",
    "Machine learning models need training data",
    "Dogs are loyal and friendly animals",
    "Cats are independent and curious pets",
]

# Encode documents into embeddings
embeddings = model.encode(documents).tolist()

# Add to Chroma collection
collection.add(
    documents=documents,
    embeddings=embeddings,
    ids=[f"doc{i}" for i in range(len(documents))],
    metadatas=[{"source": "demo", "idx": i} for i in range(len(documents))]
)

# Query with semantic search
query = "What does Machine learning need?"
q_emb = model.encode([query]).tolist()
results = collection.query(query_embeddings=q_emb, n_results=3)

# Print results
for doc, dist in zip(results["documents"][0], results["distances"][0]):
    print(f"[{dist:.3f}] {doc}")
```
