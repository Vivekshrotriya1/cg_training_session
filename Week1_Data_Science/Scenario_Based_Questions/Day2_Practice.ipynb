```python
sales_day1=np.array([120,85,60])
sales_day2=np.array([150,90,75])
total_sales=sales_day1+sales_day2
print(total_sales)
percentage_growth=((sales_day2-sales_day1)/sales_day1)*100
print(percentage_growth)
highest_growth=np.argmax(percentage_growth)
if highest_growth==0:
    print("Highest Growth=A")
elif highest_growth==1:
    print("Highest Growth=B")
else:
    print("Highest Growth=C")

```

```python
yields=np.array([[2.5,3.0,1.0],
                 [2.8,3.2,2.0],
                 [3.1,2.9,2.2],
                 [2.6,3.1,1.9]])
yields[1]=yields[1]+np.array([0.2,0.3,0.1])
print("Value added to second region")
print(yields)
yields[:,2]=yields[:,2]*1.1
print("Multiply Coulmn C by 1.1")
print(yields)
avg=np.mean(yields,axis=0)
print("highest Average yield:",np.argmax(avg))
```

```python
crops=np.array([[9000,5600,780,890],
               [7800,6500,468,7674],
               [560,877,655,2500],
               ])
crops=crops*1.10
print("Yield increased by 10%:",crops)
print("Wheat's yield:", crops[0])
print("Average yield per crop accross all seasons:",np.mean(crops,axis=1))
print("Yields greater than 3000:",crops[crops>3000])
```

```python
import numpy as np

# Create 2D array
production = np.array([
    [450, 520, 480, 500],   # Gadgets
    [300, 350, 400, 380],   # Tools
    [600, 650, 700, 720]    # Machines
])

print("Original Production Data:\n", production)
production_updated = production * 1.05
print("After 5% Increase:\n", production_updated)
gadgets = production_updated[0]
print("Gadgets Production:", gadgets)
average_production = np.mean(production_updated, axis=1)
print("Average Production per Product:", average_production)
high_production = production_updated[production_updated > 500]
print("Production values greater than 500:\n", high_production)
```
